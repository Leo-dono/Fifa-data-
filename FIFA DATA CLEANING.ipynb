```python
#define a function to remove the kg and lbs and convert all lbs to kg
def All_to_kg(x):
    if 'kg'in x:
        x = x.strip('kg')
        return x
    elif 'lbs'in x:
        x=x.strip('lbs')
        weight = round((float(x)*0.453592),2)
        return weight
```

```python
def int64_to_float64(df):
    """
    Convert all int64 columns to float64 in a Pandas DataFrame.
    
    Args:
    df (pandas.DataFrame): The DataFrame to modify.
    
    Returns:
    pandas.DataFrame: A new DataFrame with all int64 columns converted to float64.
    """
    # Find all int64 columns
    int_cols = df.select_dtypes(include=['int64']).columns
    
    # Convert int64 columns to float64
    df[int_cols] = df[int_cols].astype('float64')
    
    return df
```

```python
def convert_height_to_cm(x):
    if "'" in x:
        feet, inches = x.split("'")       
        feet = float(feet.replace("'",'') )
        inches = float(inches.replace('"','') )                       
        cm = float(feet) * 30.48 + float(inches) * 2.54
        return cm
    else :
        return float(x)
 
```

```python
df['Height_cm']=df['Height_cm'].apply(convert_height_to_cm)
df['Height_cm'].unique()
```
